# Axes , Label dan Grid

<em>Import lybrary nya</em>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<em>Import value yang akan dicari</em>

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

<em>Memplot value yang akan dihasilkan</em>

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

<em>(Axes = x dan y), (Label = Judul x dan y), Chart title = judul dari gambar yang dibuat)</em>

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

<em>Legend = Details data dari grafik yang ditampilkan</em>

```python
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best', shadow='True')
```

<em>Grid = skala yang keluar di grafik (kotak")</em>

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best', fontsize="small",shadow=True)
plt.grid()
```
