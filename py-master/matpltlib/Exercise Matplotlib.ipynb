```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title("US Technology Stocks")
plt.legend()
```

```python
plt.axis("equal")
plt.pie(revenue,labels=company, shadow=True, autopct='%1.02f%%',radius=1.5,explode=[0.1,0.1,0.1,0.1])
plt.show()
```

```python
plt.xlabel("Revenue")
plt.ylabel("Company")
plt.title("Tech Company")

plt.hist(revenue, bins=[80,100,125,200], rwidth=0.95, color='g')
```
