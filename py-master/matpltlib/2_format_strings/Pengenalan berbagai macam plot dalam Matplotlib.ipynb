# Pengenalan marker dalam matplotlib

<em>Import library matplotlib</em>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<em>Import Value yang akan dibuat</em>

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

<em>Mulai plot tabel dengan memasukkan spesifikasi<em/>

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')
```

<em>Menampilkan plot dengan gambar yang berbeda yaitu dengan menggunakan marker</em>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

<em>Menampilkan plot yang berbeda dengan menggunakan marker dan markersize</em>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

<em>Alpha property digunakan untuk mengatur transparancy dari suatu grafik</em>

```python
plt.plot(x,y,'g<',alpha=0.5) 
```
